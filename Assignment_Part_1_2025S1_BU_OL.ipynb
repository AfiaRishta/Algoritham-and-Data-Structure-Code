````python

# For: 2025-CSE2ALG(OL-1) - Algorithms and Data Structures
# Author: Afia Zahin Rishta (Student)

# Insertion Sort Lexicon and Hash Table Demo


# Define the Word class.
class Word:
    def __init__(self, word_str):
        """
        Initialize a Word object.

        Parameters:
            word_str (str): The text of the word.
        """
        self.word = word_str.lower()
        self.frequency = 1
        self.neighbours = []  # Neighbours will be added later.

    def increment_frequency(self):
        self.frequency += 1

    def add_neighbour(self, neighbour_word):
        if neighbour_word not in self.neighbours:
            self.neighbours.append(neighbour_word)

    def is_neighbour(self, other_word):
        """Return True if other_word differs by exactly one letter."""
        if len(self.word) != len(other_word):
            return False
        diff_count = 0
        for a, b in zip(self.word, other_word.lower()):
            if a != b:
                diff_count += 1
                if diff_count > 1:
                    return False
        return diff_count == 1

    def __lt__(self, other):
        return self.word < other.word

    def __str__(self):
        return f"Word('{self.word}', Freq={self.frequency})"


#  Sorting Algorithm 1: Insertion Sort
def insertion_sort_lexicon(lexicon):
    """
    Sorts a list of Word objects in place using insertion sort.

    Parameters:
        lexicon (list): A list of Word objects to be sorted alphabetically.
    """
    n = len(lexicon)
    for i in range(1, n):
        key = lexicon[i]
        j = i - 1
        while j >= 0 and lexicon[j] > key:
            lexicon[j + 1] = lexicon[j]
            j -= 1
        lexicon[j + 1] = key


# Hash Table with Chaining Implementation
class HTNode:
    def __init__(self, word_obj):
        """
        Node for the hash table chain.

        Parameters:
            word_obj (Word): The Word object to store.
        """
        self.data = word_obj
        self.next = None


class HashTable:
    def __init__(self, size=101):
        """
        Initialize the hash table with a specified number of buckets.

        Parameters:
            size (int): Number of buckets.
        """
        self.size = size
        self.buckets = [None] * self.size

    def hash_function(self, key):
        """
        Computes the bucket index for a given key.

        Parameters:
            key (str): The word (key) to hash.

        Returns:
            int: Bucket index.
        """
        return hash(key) % self.size

    def insert(self, word_obj):
        """
        Inserts a Word object using chaining.

        If the word already exists, it updates the frequency.
        """
        key = word_obj.word
        index = self.hash_function(key)
        current = self.buckets[index]

        if current is None:
            self.buckets[index] = HTNode(word_obj)
            return
        else:
            while current:
                if current.data.word == key:
                    # Update frequency if already exists.
                    current.data.frequency = word_obj.frequency
                    return
                if current.next is None:
                    break
                current = current.next
            current.next = HTNode(word_obj)

    def search(self, key):
        """
        Searches for a Word object by its text.

        Returns the Word object if found, otherwise None.
        """
        index = self.hash_function(key)
        current = self.buckets[index]
        while current:
            if current.data.word == key:
                return current.data
            current = current.next
        return None

    def delete(self, key):
        """
        Deletes the Word object with the given key.

        Returns True if deletion was successful, False otherwise.
        """
        index = self.hash_function(key)
        current = self.buckets[index]
        previous = None
        while current:
            if current.data.word == key:
                if previous is None:
                    self.buckets[index] = current.next
                else:
                    previous.next = current.next
                return True
            previous = current
            current = current.next
        return False

    def display(self):
        """
        Displays each bucket and its chain of Word objects.
        """
        for idx, node in enumerate(self.buckets):
            if node:
                chain = []
                current = node
                while current:
                    chain.append(str(current.data))
                    current = current.next
                print(f"Bucket {idx}: " + " -> ".join(chain))
            else:
                print(f"Bucket {idx}: Empty")



def main():
    # Create a sample lexicon (unsorted list) of Word objects.
    sample_lexicon = [
        Word("banana"),
        Word("apple"),
        Word("cherry"),
        Word("date"),
        Word("elderberry")
    ]
    # Adjust frequencies for demonstration
    sample_lexicon[0].frequency = 2  # banana
    sample_lexicon[1].frequency = 3  # apple

    print("Original Lexicon:")
    for w in sample_lexicon:
        print(w)

    # Test Insertion Sort
    lexicon_insertion = sample_lexicon.copy()
    insertion_sort_lexicon(lexicon_insertion)
    print("\nLexicon Sorted using Insertion Sort:")
    for w in lexicon_insertion:
        print(w)

    #  Hash Table Demo
    print("\nConstructing Hash Table (for Word objects)...")
    ht = HashTable(size=17)  # Using a small size for demonstration.
    for word_obj in sample_lexicon:
        ht.insert(word_obj)

    print("\nHash Table Contents:")
    ht.display()

    # Test search in the hash table.
    key_to_search = "cherry"
    found = ht.search(key_to_search)
    if found:
        print(f"\nSearch: Found {found}")
    else:
        print(f"\nSearch: '{key_to_search}' not found.")

    # Test deletion.
    key_to_delete = "banana"
    if ht.delete(key_to_delete):
        print(f"\nDeletion: '{key_to_delete}' deleted successfully.")
    else:
        print(f"\nDeletion: '{key_to_delete}' not found.")

    print("\nHash Table Contents After Deletion:")
    ht.display()

if __name__ == "__main__":
    main()
````

````python

# For: 2025-CSE2ALG(OL-1) - Algorithms and Data Structures
# Author: Afia Zahin Rishta (Student)

#  Hybrid Merge-Insertion Sort Lexicon and Hash Table Demo


THRESHOLD = 10  # Threshold for switching to insertion sort in the hybrid algorithm

# Define the Word class.
class Word:
    def __init__(self, word_str):
        """
        Initialize a Word object.

        Parameters:
            word_str (str): The text of the word.
        """
        self.word = word_str.lower()
        self.frequency = 1
        self.neighbours = []  # This can be populated later.

    def increment_frequency(self):
        self.frequency += 1

    def add_neighbour(self, neighbour_word):
        if neighbour_word not in self.neighbours:
            self.neighbours.append(neighbour_word)

    def is_neighbour(self, other_word):
        if len(self.word) != len(other_word):
            return False
        diff_count = 0
        for a, b in zip(self.word, other_word.lower()):
            if a != b:
                diff_count += 1
                if diff_count > 1:
                    return False
        return diff_count == 1

    def __lt__(self, other):
        return self.word < other.word

    def __str__(self):
        return f"Word('{self.word}', Freq={self.frequency})"

# Sorting Algorithm 2: Hybrid Merge-Insertion Sort
def hybrid_sort_lexicon(lexicon):
    """
    Sorts a list of Word objects in place using a hybrid approach that uses merge sort
    for larger subarrays and switches to insertion sort for small subarrays.

    Parameters:
        lexicon (list): A list of Word objects to be sorted alphabetically.
    """
    def hybrid_sort_helper(arr, left, right):
        if right - left + 1 <= THRESHOLD:
            # Use insertion sort on small subarrays.
            for i in range(left + 1, right + 1):
                key = arr[i]
                j = i - 1
                while j >= left and arr[j] > key:
                    arr[j + 1] = arr[j]
                    j -= 1
                arr[j + 1] = key
            return

        mid = (left + right) // 2
        hybrid_sort_helper(arr, left, mid)
        hybrid_sort_helper(arr, mid + 1, right)
        merge(arr, left, mid, right)

    def merge(arr, left, mid, right):
        """
        Merges two sorted subarrays into a single sorted subarray.
        """
        L = arr[left:mid + 1]
        R = arr[mid + 1:right + 1]
        i = 0  # Pointer for L.
        j = 0  # Pointer for R.
        k = left  # Pointer for merged array.

        while i < len(L) and j < len(R):
            if L[i] < R[j]:
                arr[k] = L[i]
                i += 1
            else:
                arr[k] = R[j]
                j += 1
            k += 1

        # Copy the remaining elements from L.
        while i < len(L):
            arr[k] = L[i]
            i += 1
            k += 1

        # Copy the remaining elements from R.
        while j < len(R):
            arr[k] = R[j]
            j += 1
            k += 1

    if lexicon:
        hybrid_sort_helper(lexicon, 0, len(lexicon) - 1)

#  Hash Table with Chaining
class HTNode:
    def __init__(self, word_obj):
        self.data = word_obj
        self.next = None

class HashTable:
    def __init__(self, size=101):
        self.size = size
        self.buckets = [None] * self.size

    def hash_function(self, key):
        return hash(key) % self.size

    def insert(self, word_obj):
        key = word_obj.word
        index = self.hash_function(key)
        current = self.buckets[index]
        if current is None:
            self.buckets[index] = HTNode(word_obj)
            return
        else:
            while current:
                if current.data.word == key:
                    current.data.frequency = word_obj.frequency
                    return
                if current.next is None:
                    break
                current = current.next
            current.next = HTNode(word_obj)

    def search(self, key):
        index = self.hash_function(key)
        current = self.buckets[index]
        while current:
            if current.data.word == key:
                return current.data
            current = current.next
        return None

    def delete(self, key):
        index = self.hash_function(key)
        current = self.buckets[index]
        previous = None
        while current:
            if current.data.word == key:
                if previous is None:
                    self.buckets[index] = current.next
                else:
                    previous.next = current.next
                return True
            previous = current
            current = current.next
        return False

    def display(self):
        for idx, node in enumerate(self.buckets):
            if node:
                chain = []
                current = node
                while current:
                    chain.append(str(current.data))
                    current = current.next
                print(f"Bucket {idx}: " + " -> ".join(chain))
            else:
                print(f"Bucket {idx}: Empty")

def main():
    # Create a sample lexicon (unsorted list) of Word objects.
    sample_lexicon = [
        Word("banana"),
        Word("apple"),
        Word("cherry"),
        Word("date"),
        Word("elderberry")
    ]
    # Adjust frequencies arbitrarily for demonstration.
    sample_lexicon[0].frequency = 2   # banana
    sample_lexicon[1].frequency = 3   # apple

    print("Original Lexicon:")
    for w in sample_lexicon:
        print(w)

    # Test Hybrid Merge-Insertion Sort
    lexicon_hybrid = sample_lexicon.copy()
    hybrid_sort_lexicon(lexicon_hybrid)
    print("\nLexicon Sorted using Hybrid Merge-Insertion Sort:")
    for w in lexicon_hybrid:
        print(w)

    # Hash Table Demo
    print("\nConstructing Hash Table (for Word objects)...")
    ht = HashTable(size=17)  # Use a small table size for demonstration.
    for word_obj in sample_lexicon:
        ht.insert(word_obj)

    print("\nHash Table Contents:")
    ht.display()

    # Test search in the hash table.
    key_to_search = "cherry"
    found = ht.search(key_to_search)
    if found:
        print(f"\nSearch: Found {found}")
    else:
        print(f"\nSearch: '{key_to_search}' not found.")

    # Test deletion.
    key_to_delete = "banana"
    if ht.delete(key_to_delete):
        print(f"\nDeletion: '{key_to_delete}' deleted successfully.")
    else:
        print(f"\nDeletion: '{key_to_delete}' not found.")

    print("\nHash Table Contents After Deletion:")
    ht.display()

if __name__ == "__main__":
    main()
````

````python

# For: 2025-CSE2ALG(OL-1) - Algorithms and Data Structures
# Author: Afia Zahin Rishta (Student)

# Step:1: Lexicon to File Output

# Custom Insertion Sort for a list of strings (for sorting keys and neighbours)
def insertion_sort_strings(lst):
    """
    Sorts a list of strings in place using insertion sort.

    Parameters:
        lst (list of str): The list of strings to sort.
    """
    n = len(lst)
    for i in range(1, n):
        key = lst[i]
        j = i - 1
        while j >= 0 and lst[j] > key:
            lst[j + 1] = lst[j]
            j -= 1
        lst[j + 1] = key

# Word class definition
class Word:
    def __init__(self, word_str):
        """
        Initialize a Word object.

        Parameters:
            word_str (str): The word's text.
        Instance Variables:
            word        - the spelling of the word (stored in lowercase)
            frequency   - count of occurrences (starts at 1)
            neighbours  - list of neighbour words (empty initially)
        """
        self.word = word_str.lower()
        self.frequency = 1
        self.neighbours = []

    def increment_frequency(self):
        """Increment the frequency by 1."""
        self.frequency += 1

    def add_neighbour(self, neighbour_word):
        """
        Adds a neighbour word if not already in the list.

        Parameters:
            neighbour_word (str): A neighbour word.
        """
        if neighbour_word not in self.neighbours:
            self.neighbours.append(neighbour_word)

    def is_neighbour(self, other_word):
        """
        Checks if another word (string) is a neighbour.
        Two words are neighbours if they are of the same length
        and differ by exactly one letter.

        Parameters:
            other_word (str): The word to compare.
        Returns:
            bool: True if neighbour, otherwise False.
        """
        if len(self.word) != len(other_word):
            return False
        diff_count = 0
        for a, b in zip(self.word, other_word.lower()):
            if a != b:
                diff_count += 1
                if diff_count > 1:
                    return False
        return diff_count == 1

    def __lt__(self, other):
        """Less-than operator: sorts Word objects by their word text."""
        return self.word < other.word

    def __str__(self):
        """
        Returns the string representation in the required format:
            word frequency [neighbour1, neighbour2, ...]
        """
        return f"{self.word} {self.frequency} {self.neighbours}"

def add_neighbours(lexicon):
    """
    Populates each Word object's neighbours list with words from the lexicon
    that are of the same length and differ by exactly one letter.
    The neighbours list is then sorted using our custom insertion sort.

    Parameters:
        lexicon (dict): Dictionary mapping word strings to Word objects.
    """
    words = list(lexicon.values())
    for word_obj in words:
        # Clear previous neighbours (if any)
        word_obj.neighbours = []
        for candidate in words:
            if word_obj.word != candidate.word and len(word_obj.word) == len(candidate.word):
                if word_obj.is_neighbour(candidate.word):
                    word_obj.neighbours.append(candidate.word)
        # Sort the neighbours list alphabetically using our custom insertion sort.
        insertion_sort_strings(word_obj.neighbours)

def write_data(lexicon, filename):
    """
    Writes all words in the lexicon to a text file.
    Each line is in the format:
        word frequency [neighbour1, neighbour2, ...]
    The words are written in alphabetical order.

    Parameters:
        lexicon (dict): Dictionary mapping word strings to Word objects.
        filename (str): The output file name (e.g., "out.txt").
    """
    try:
        # Extract keys and sort them
        keys = list(lexicon.keys())
        insertion_sort_strings(keys)
        with open(filename, 'w') as outfile:
            for key in keys:
                word_obj = lexicon[key]
                # Construct the output line following the required format.
                line = f"{word_obj.word} {word_obj.frequency} {word_obj.neighbours}\n"
                outfile.write(line)
        print(f"Lexicon successfully written to '{filename}'.")
    except Exception as e:
        print(f"Error writing to file {filename}: {e}")

def demo_lexicon_to_file():
    """
    Demonstrates the lexicon file writing.
    Constructs a sample lexicon, populates neighbours,
    and writes the lexicon to a file called 'out.txt'.
    """
    lexicon_dict = {
        "apple": Word("apple"),
        "about": Word("about"),
        "access": Word("access"),
        "acknowledged": Word("acknowledged"),
        "action": Word("action"),
        "advanced": Word("advanced"),
        "all": Word("all"),
        "allnew": Word("allnew"),
        "also": Word("also"),
        "and": Word("and"),
        "answer": Word("answer")
    }
    # Simulate frequency values for demonstration.
    lexicon_dict["apple"].frequency = 3
    lexicon_dict["and"].frequency = 10

    # Populate neighbours (each Word is compared with every other Word).
    add_neighbours(lexicon_dict)
    # Write the lexicon to "out.txt".
    write_data(lexicon_dict, "out.txt")

# Step:2 : Sensor Data Statistical Calculations

class SensorRecord:
    def __init__(self, sensor_id, temperature, pressure, humidity, timestamp):
        """
        Initializes a sensor record.

        Parameters:
            sensor_id (str): Unique sensor ID.
            temperature (str): Temperature reading.
            pressure (str): Pressure reading.
            humidity (str): Humidity value.
            timestamp (str): Timestamp of the record.
        """
        self.sensor_id = sensor_id
        self.temperature = temperature
        self.pressure = pressure
        self.humidity = humidity
        self.timestamp = timestamp

    def __str__(self):
        return (f"SensorRecord(ID={self.sensor_id}, Temp={self.temperature}, "
                f"Pressure={self.pressure}, Humidity={self.humidity}, Time={self.timestamp})")

def compute_sensor_statistics(records, attr):
    """
    Computes the minimum, maximum, and average for a given numeric attribute from sensor records.

    Parameters:
        records (list): A list of SensorRecord objects.
        attr (str): The attribute name to compute (e.g., "temperature" or "pressure").

    Returns:
        tuple: (min_value, max_value, average) as floats; returns (None, None, None) if no valid values.
    """
    if not records:
        return None, None, None
    total = 0.0
    count = 0
    min_val = None
    max_val = None
    for rec in records:
        try:
            value = float(getattr(rec, attr))
        except (ValueError, AttributeError):
            continue
        if min_val is None or value < min_val:
            min_val = value
        if max_val is None or value > max_val:
            max_val = value
        total += value
        count += 1
    if count == 0:
        return None, None, None
    average = total / count
    return min_val, max_val, average

def demo_statistics():
    """
    Demonstrates statistical calculations on a sample list of SensorRecord objects.
    Prints minimum, maximum, and average values for temperature and pressure.
    """
    sample_records = [
        SensorRecord("1", "23.5", "1013", "45", "2025-04-12 08:00:00"),
        SensorRecord("2", "24.1", "1012", "50", "2025-04-12 08:05:00"),
        SensorRecord("3", "22.8", "1014", "47", "2025-04-12 08:10:00")
    ]
    min_temp, max_temp, avg_temp = compute_sensor_statistics(sample_records, "temperature")
    print("Temperature Statistics:")
    print(f"  Minimum: {min_temp}")
    print(f"  Maximum: {max_temp}")
    print(f"  Average: {avg_temp}")

    min_press, max_press, avg_press = compute_sensor_statistics(sample_records, "pressure")
    print("\nPressure Statistics:")
    print(f"  Minimum: {min_press}")
    print(f"  Maximum: {max_press}")
    print(f"  Average: {avg_press}")

# Execute Both Parts

def main():
    print("Lexicon to File Output")
    # Run the lexicon-to-file demonstration
    demo_lexicon_to_file()

    print("Sensor Data Statistics")
    demo_statistics()

if __name__ == '__main__':
    main()
````
